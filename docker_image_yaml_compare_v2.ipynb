```python
import yaml
import re

# Données directement fournies dans le script
donnees_sortie = """
Répertoire: accuweather
    image: gitlab-daconline.dactechnologies.net:5050/microservice/accuweather-api:0.1.0-rc
------------------------------
Répertoire: backadmin
    image: gitlab-daconline.dactechnologies.net:5050/cagbolossou/chatbot_backoffice_admin:0.1.0-rc
------------------------------
Répertoire: currency
    image: gitlab-daconline.dactechnologies.net:5050/cagbolossou/currency-new-api:dev
------------------------------
Répertoire: dacpharma-admin
    image: gitlab-daconline.dactechnologies.net:5050/pharmacy/f_pharma_admin_tg:prod
------------------------------
Répertoire: dacpharma-api-rc1
    image: gitlab-daconline.dactechnologies.net:5050/pharmacy/pharmacie-api:prod
------------------------------
Répertoire: dacpharma-pool
    image: gitlab-daconline.dactechnologies.net:5050/pharmacy/d_pharmacy_pools:prod
------------------------------
Répertoire: flooz_callback
    image: gitlab-daconline.dactechnologies.net:5050/microservice/flooz-callback:prod
------------------------------
Répertoire: frontadmin
    image: gitlab-daconline.dactechnologies.net:5050/dacbot/chatbot_banking_backoffice_admin:0.1.0-rc
------------------------------
Répertoire: gotenberg
    image: thecodingmachine/gotenberg:7
------------------------------
Répertoire: intltel
    image: gitlab-daconline.dactechnologies.net:5050/mtchakponou/dac-intltel:0.1.0-rc
------------------------------
Répertoire: lonato
    image: gitlab-daconline.dactechnologies.net:5050/eavlah/lonato-microservice-api:0.1.0-rc
------------------------------
Répertoire: mongo
    image: mongo:4.4
------------------------------
Répertoire: otp
    image: gitlab-daconline.dactechnologies.net:5050/microservice/dac-otp:prod
------------------------------
Répertoire: pharmainfo-prod
    image: gitlab-daconline.dactechnologies.net:5050/pharmacy/pharmacie-front:0.1.0-rc
------------------------------
Répertoire: pharma-webhook-rc1
    image: gitlab-daconline.dactechnologies.net:5050/pharmacy/pharmacie-webhook:prod
    image: redis:4-alpine
------------------------------
Répertoire: tmoney_callback
    image: gitlab-daconline.dactechnologies.net:5050/microservice/dac-callback:prod
------------------------------
Répertoire: webhook-tester
    image: gitlab.dactechnologies.net:5050/eavlah/chatbot-banking-webhook:0.1.0-rc
    image: redis:4-alpine
"""

# Données fournies dans le script pour les fichiers YAML
donnees_yaml = {
    "delivery": [
        "delivery-api",
        "delivery-organisation-web-app-v2",
        "delivery-admin-web-app"
    ],
    "dacpharmapro": [
        "dacpharmapro_saml_project",
        "d_pharmacy_pools_flutter_v2"
    ],
    "generation_de_tour_de_garde": [
        "pharmacy_pools_auto_gen_api"
    ],
    "dacpharma": [
        "pharmacie-api",
        "f_pharma_admin_tg",
        "d_pharmacy_pools_flutter_v2",
        "pharmacie-webhook"
    ],
    "microservices": [
        "dac-mirror-api",
        "api-manager",
        "flooz-callback",
        "dac-otp",
        "new-tmoney-callback",
        "dac-intltel"
    ],
    "dacpay": [
        "dac_dacpaye_server"
    ],
    "dacbot": [
        "Currency-api",
        "accuweather-api",
        "lonato-microservice-api",
        "chatbot_banking_backoffice_admin",
        "chatbot_backoffice_admin",
        "chatbot_partenaire_api_microfina"
    ],
    "marketplace": [
        "ebeez_api",
        "ebeez_dashboard"
    ]
}

def formater_donnees_sortie(donnees_sortie):
    donnees_structurees = {}
    repertoire_en_cours = None
    lignes = donnees_sortie.split("\n")
    
    for ligne in lignes:
        if ligne.startswith("Répertoire:"):
            repertoire_en_cours = ligne.split(": ")[1].strip()
            donnees_structurees[repertoire_en_cours] = []
        elif ligne.startswith("    image:"):
            image = ligne.split(": ")[1].strip()
            donnees_structurees[repertoire_en_cours].append(image)

    # Formater les données pour les rendre similaires aux données YAML
    donnees_formatees = {}
    for fichier, parties_urls in donnees_structurees.items():
        parties_formatees = []
        for partie_url in parties_urls:
            # Remplacer les caractères spéciaux par des underscores et mettre en minuscules
            partie_url_formatee = re.sub(r'[^a-zA-Z0-9]', '_', partie_url.lower())
            parties_formatees.append(partie_url_formatee)
        donnees_formatees[fichier] = parties_formatees
    
    print("\nDonnées formatées de sortie (sous la même forme que données YAML) :")
    print(donnees_formatees)
    
    return donnees_formatees

def formater_parties_urls(donnees_yaml):
    parties_urls_formatees = {}
    
    for fichier, parties_urls in donnees_yaml.items():
        parties_urls_formatees[fichier] = set(
            partie.lower().replace(" ", "") for partie in parties_urls
        )
    
    print("\nParties d'URLs formatées :")
    print(parties_urls_formatees)
    
    return parties_urls_formatees

def comparer_repertoires(donnees_structurees, parties_urls_formatees):
    resultats = []
    
    for fichier, parties_urls in parties_urls_formatees.items():
        parties_repertoires_structurees = set(
            partie.lower().replace(" ", "") for repertoire in donnees_structurees.keys() for partie in repertoire.split('/')
        )

        if set(parties_urls).intersection(parties_repertoires_structurees):
            resultats.append({
                "fichier_yaml": fichier,
                "repertoires_similaires": list(donnees_structurees.keys())
            })
    
    print("\nRésultats de la comparaison des répertoires :")
    print(resultats)
    
    return resultats

def afficher_resultats(resultats):
    for resultat in resultats:
        print(f"\nRépertoires similaires dans {resultat['fichier_yaml']} :")
        for repertoire_similaire in resultat['repertoires_similaires']:
            print(f"  - {repertoire_similaire}")

donnees_structurees = formater_donnees_sortie(donnees_sortie)
parties_urls_formatees = formater_parties_urls(donnees_yaml)
resultats = comparer_repertoires(donnees_structurees, parties_urls_formatees)
afficher_resultats(resultats)
```
