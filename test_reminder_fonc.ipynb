I understand that you want to integrate the script for updating and deploying your Django project with the systemd configuration you've provided. Let's modify the systemd configuration files to fit your project.

1. **Update gunicorn.service:**

   Modify the `/etc/systemd/system/gunicorn.service` file with the following changes:

   ```ini
   [Unit]
   Description=gunicorn daemon for YourProject
   Requires=gunicorn.socket
   After=network.target

   [Service]
   Type=notify
   User=someuser
   Group=someuser
   RuntimeDirectory=gunicorn
   WorkingDirectory=/path/to/your/project
   ExecStart=/path/to/your/virtualenv/bin/gunicorn your_project.wsgi:application
   ExecReload=/bin/kill -s HUP $MAINPID
   KillMode=mixed
   TimeoutStopSec=5
   PrivateTmp=true

   [Install]
   WantedBy=multi-user.target
   ```

   Update the `User`, `WorkingDirectory`, and `ExecStart` fields to reflect your specific project details.

2. **Update gunicorn.socket:**

   Modify the `/etc/systemd/system/gunicorn.socket` file with the following changes:

   ```ini
   [Unit]
   Description=gunicorn socket for YourProject

   [Socket]
   ListenStream=/run/gunicorn.sock
   SocketUser=www-data

   [Install]
   WantedBy=sockets.target
   ```

   Ensure that `SocketUser` matches the user that your Nginx service runs as. If you are using a different user, adjust it accordingly.

3. **Enable and start the systemd services:**

   Run the following commands to enable and start the socket:

   ```bash
   sudo systemctl enable --now gunicorn.socket
   ```

4. **Update Nginx configuration:**

   Modify your Nginx configuration file, typically located at `/etc/nginx/nginx.conf` or in a site-specific configuration file in the `/etc/nginx/sites-available/` directory:

   ```nginx
   user www-data;
   ...
   http {
       server {
           listen          8000;
           server_name     your_domain_or_ip;

           location / {
               proxy_pass http://unix:/run/gunicorn.sock;
               proxy_set_header Host $host;
               proxy_set_header X-Real-IP $remote_addr;
               proxy_set_header X-Forwarded-For $proxy_add_x_forwarded_for;
           }
       }
   }
   ...
   ```

   Replace `your_domain_or_ip` with your actual domain or IP address.

5. **Enable and start Nginx:**

   Run the following commands to enable and start Nginx:

   ```bash
   sudo systemctl enable nginx.service
   sudo systemctl start nginx
   ```
