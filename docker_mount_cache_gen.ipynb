```python
import os
import shutil

def copy_selected_files(source, destination, selected_files):
    # Copier uniquement les fichiers spécifiés depuis le répertoire source vers le répertoire destination
    for item in os.listdir(source):
        source_path = os.path.join(source, item)
        destination_path = os.path.join(destination, item)

        if os.path.isdir(source_path):
            # Ignorer le répertoire .git
            if item == ".git":
                continue
            
            # Créer le répertoire de destination s'il n'existe pas
            if not os.path.exists(destination_path):
                os.makedirs(destination_path)

            # Copier le contenu du sous-répertoire récursivement
            copy_selected_files(source_path, destination_path, selected_files)
        else:
            # Vérifier si le fichier source est dans la liste des fichiers sélectionnés
            if item in selected_files:
                # Copier le fichier en écrasant celui déjà présent
                shutil.copy2(source_path, destination_path)

def main():
    # Spécifier les répertoires locaux et en ligne
    local_directory = "/home/julien/Téléchargements/migration_upd/local_version"
    online_directory = "/home/julien/Téléchargements/migration_upd/online_version"

    # Liste des fichiers à copier
    selected_files = ["Dockerfile", ".gitlab-ci.yml"]

    # Liste des répertoires dans le répertoire en ligne
    online_subdirectories = [d for d in os.listdir(online_directory) if os.path.isdir(os.path.join(online_directory, d))]

    # Parcourir les répertoires et effectuer la copie des fichiers sélectionnés
    for subdir in online_subdirectories:
        online_subdir_path = os.path.join(online_directory, subdir)
        local_subdir_path = os.path.join(local_directory, subdir)

        # Vérifier si le répertoire local existe
        if os.path.exists(local_subdir_path):
            # Copier les fichiers sélectionnés du répertoire en ligne vers le répertoire local
            copy_selected_files(online_subdir_path, local_subdir_path, selected_files)
        else:
            print(f"Le répertoire local {subdir} n'existe pas.")

if __name__ == "__main__":
    main()
```
