```python
    """
    Génère des scripts de mise à jour et des fichiers Docker Compose pour chaque projet dans la structure de groupes fournie.

    Args:
    groups (dict): Un dictionnaire représentant la structure des groupes et projets.

    Example:
    groups = {
        "Group1": {
            "Project1": "https://gitlab.example.com/group1/project1",
            "Project2": "https://gitlab.example.com/group1/project2",
        },
        "Group2": {
            "Project3": "https://gitlab.example.com/group2/project3",
            "Project4": "https://gitlab.example.com/group2/project4",
        },
    }
    """
import os

def generate_scripts_and_compose_files(groups):

    for group, projects in groups.items():
        # Créer un répertoire pour le groupe
        group_directory = f"./{group.lower()}"
        os.makedirs(group_directory, exist_ok=True)

        for project, url in projects.items():
            # Utiliser la partie finale de l'URL comme nom du projet
            project_name = url.split("/")[-1]

            # Générer le nom du script et son contenu
            script_name = f"update-{project_name.lower().replace('_', '-')}.sh"
            script_content = f"""#!/bin/bash
# Script de mise à jour pour {project_name}
# Url git {url}
# Groupe {group_directory}

# Redémarrer le projet
docker compose pull
docker compose up -d --force-recreate
"""

            # Générer le nom du fichier Docker Compose et son contenu
            compose_file_name = "compose.yml"
            compose_content = f"# {url}"

            # Écrire le script de mise à jour
            script_path = os.path.join(group_directory, script_name)
            with open(script_path, "w") as script_file:
                script_file.write(script_content)

            # Écrire le fichier Docker Compose
            compose_path = os.path.join(group_directory, compose_file_name)
            with open(compose_path, "w") as compose_file:
                compose_file.write(compose_content)

            # Afficher des informations sur les fichiers générés
            print(f"Script généré: {script_path}")
            print(f"Fichier Docker Compose généré: {compose_path}\n")

# Définir la structure des groupes et projets
groups = {
    "Delivery": {
        "API": "https://gitlab-daconline.dactechnologies.net/dac_delivery/delivery-api",
        "Interface Organisations Front": "https://gitlab-daconline.dactechnologies.net/dac_delivery/delivery-organisation-web-app-v2",
        "Interface admin": "https://gitlab-daconline.dactechnologies.net/dac_delivery/delivery-admin-web-app",
    },
    "DacPharmaPro": {
        "API": "https://gitlab-daconline.dactechnologies.net/pharmacy/dacpharmapro_saml_project",
        "Front end": "https://gitlab-daconline.dactechnologies.net/pharmacy/d_pharmacy_pools_flutter_v2",
    },
    "Génération de tour de garde": {
        "API": "https://gitlab-daconline.dactechnologies.net/pharmacy/pharmacy_pools_auto_gen_api",
    },
    "DacPharma": {
        "API": "https://gitlab-daconline.dactechnologies.net/pharmacy/pharmacie-api",
        "Interface Admin": "https://gitlab-daconline.dactechnologies.net/pharmacy/f_pharma_admin_tg",
        "Webhook": "https://gitlab-daconline.dactechnologies.net/pharmacy/pharmacie-webhook",
        "Statistiques": "https://gitlab-daconline.dactechnologies.net/cagbolossou/dac_stats_front",
        "API Statistiques": "https://gitlab-daconline.dactechnologies.net/mtchakponou/global-stat",
    },
    "Microservices": {
        "API calcule disances": "https://gitlab-daconline.dactechnologies.net/microservice/dac-mirror-api",
        "API Manager": "https://gitlab-daconline.dactechnologies.net/microservice/api-manager",
        "FloozCallback": "https://gitlab-daconline.dactechnologies.net/microservice/flooz-callback",
        "OTP": "https://gitlab-daconline.dactechnologies.net/microservice/dac-otp",
        "TMONEY CALLBACK": "https://gitlab-daconline.dactechnologies.net/microservice/new-tmoney-callback",
        "Vérificateur Tel Number": "https://gitlab-daconline.dactechnologies.net/mtchakponou/dac-intltel",
    },
    "DacPay": {
        "API": "https://gitlab-daconline.dactechnologies.net/microservice/dac_dacpaye_server",
    },
    "Dacbot": {
        "Currency": "https://gitlab-daconline.dactechnologies.net/microservice/Currency-api",
        "AccuWeather": "https://gitlab-daconline.dactechnologies.net/microservice/accuweather-api",
        "Lonato": "https://gitlab-daconline.dactechnologies.net/eavlah/lonato-microservice-api",
        "Front admin": "https://gitlab-daconline.dactechnologies.net/dacbot/chatbot_banking_backoffice_admin",
        "Back Admin": "https://gitlab-daconline.dactechnologies.net/cagbolossou/chatbot_backoffice_admin",
        "Microfina": "https://gitlab-daconline.dactechnologies.net/cagbolossou/chatbot_partenaire_api_microfina",
    },
    "Marketplace": {
        "API": "https://gitlab-daconline.dactechnologies.net/e_business/ebeez_api",
        "Frontend": "https://gitlab-daconline.dactechnologies.net/e_business/ebeez_dashboard",
    },
}

# Générer les scripts et les fichiers Docker Compose
generate_scripts_and_compose_files(groups)
```
