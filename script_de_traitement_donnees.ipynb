```python
#!/usr/bin/env python
# coding: utf-8

import pandas as pd
import matplotlib.pyplot as plt
import statistics

def load_and_plot_csv(file_path, columns_to_convert):
    """
    Charge le fichier CSV, convertit les colonnes sélectionnées en float64
    et affiche l'évolution des colonnes en fonction du temps.
    """
    df = pd.read_csv(file_path, sep=';')

    for column in columns_to_convert:
        df[column] = df[column].astype(float)

    df.plot(x='temps', y=columns_to_convert, kind='line')
    plt.xlabel('Temps')
    plt.ylabel('Valeurs')
    plt.title('Évolution des Colonnes en Fonction du Temps')
    plt.show()

    calculate_statistics(df[columns_to_convert])

def calculate_statistics(df):
    """
    Calcule et affiche les moyennes, médianes, modes et sommes des colonnes sélectionnées.
    """
    moyennes = df.mean()
    print("Moyennes des Colonnes:")
    print(moyennes)

    medianes = df.median()
    print("\nMédianes des Colonnes:")
    print(medianes)

    modes = df.mode().iloc[0]
    print("\nModes des Colonnes:")
    print(modes)

    sommes = df.sum()
    print("\nSomme de chaque Colonne:")
    print(sommes)

def main():
    data_files = ['data_unit.csv', 'data_guni.csv', 'data_uvi.csv']
    columns_to_convert = ['temps_reponse']  # Remplacez par les noms de vos colonnes

    for file in data_files:
        load_and_plot_csv(file, columns_to_convert)

    container_files = ['container_stats_unit.csv', 'container_stats_guni.csv', 'container_stats_uvi.csv']
    container_columns_to_convert = ['CPU %', 'MEM USAGE (MB)', 'MEM %', 'PIDS']

    for container_file in container_files:
        load_and_plot_csv(container_file, container_columns_to_convert)

if __name__ == "__main__":
    main()
```
