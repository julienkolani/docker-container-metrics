```python
#!/usr/bin/env python
# coding: utf-8

import requests
import time
import pandas as pd

def monitor_urls(urls, refresh_interval, duration):
    """
    Surveille les URL spécifiées pour une certaine durée avec un intervalle de rafraîchissement.
    Enregistre les données dans un fichier CSV.
    """
    data = []

    start_time = time.time()

    while time.time() - start_time < duration:
        print(time.time() - start_time)
        for url in urls:
            try:
                response = perform_request(url)
                record_data(data, url, response)
            except Exception as e:
                print(f"Une erreur s'est produite pour l'URL {url}: {str(e)}")

        time.sleep(refresh_interval)

    save_to_csv(data)

def perform_request(url):
    """
    Effectue une requête GET sur l'URL et renvoie la réponse.
    """
    start_time_request = time.time()
    response = requests.get(url)
    end_time_request = time.time()

    return {
        "url": url,
        "code_statut": response.status_code,
        "temps_reponse": end_time_request - start_time_request
    }

def record_data(data, url, response):
    """
    Ajoute les données à la liste.
    """
    data.append({
        "temps": time.time() - start_time,
        "url": url,
        "code_statut": response["code_statut"],
        "temps_reponse": response["temps_reponse"]
    })

def save_to_csv(data):
    """
    Convertit les données en fichier CSV.
    """
    df = pd.DataFrame(data)
    df.to_csv("data_unit.csv", index=False, sep=";")

def main():
    urls_to_monitor = ["http://172.17.0.9/siteunit.com/"]
    refresh_interval = 1
    duration = 604

    monitor_urls(urls_to_monitor, refresh_interval, duration)

if __name__ == "__main__":
    main()
```
